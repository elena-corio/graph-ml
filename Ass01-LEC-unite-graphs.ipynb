## 2.1. Import the OBJ file

```python
rooms01 = Topology.ByOBJPath(r"C:\Users\ecori\OneDrive\Documenti\MaCAD\Graph ML -- DOCUMENTS\Ass01\ap01_rooms.obj", selfMerge=True)
rooms02 = Topology.ByOBJPath(r"C:\Users\ecori\OneDrive\Documenti\MaCAD\Graph ML -- DOCUMENTS\Ass01\ap02_rooms.obj", selfMerge=True)
doors01 = Topology.ByOBJPath(r"C:\Users\ecori\OneDrive\Documenti\MaCAD\Graph ML -- DOCUMENTS\Ass01\ap01_doors.obj", selfMerge=True)
doors02 = Topology.ByOBJPath(r"C:\Users\ecori\OneDrive\Documenti\MaCAD\Graph ML -- DOCUMENTS\Ass01\ap02_doors.obj", selfMerge=True)
print(rooms01)
print(rooms02)
print(doors01)
print(doors02)
```

## 2.2. Create and Clean Cell Complex

```python
# =========== APARTMENT 01 ============
ap01 = CellComplex.ByCells(rooms01)
ap01 = Topology.RemoveCoplanarFaces(ap01)
ap01 = Topology.RemoveCollinearEdges(ap01)
ap01 = Topology.AddApertures(ap01, doors01, subTopologyType="face")

# =========== APARTMENT 02 ============
ap02 = CellComplex.ByCells(rooms02)
ap02 = Topology.RemoveCoplanarFaces(ap02)
ap02 = Topology.RemoveCollinearEdges(ap02)
ap02 = Topology.AddApertures(ap02, doors02, subTopologyType="face")
```

## 2.3. Show the geometry

```python
#Note: apertures are not automatically shown even if assigned to house. You need to explicitly add them to the list of faces to be shown.
Topology.Show(ap01,ap02,
              backgroundColor="white",
              width=500,
              height=500,
              renderer=renderer)
```

## 3.1. Derive Primal Graph

```python
# =========== APARTMENT 01 ============
# Cluster from cell complex and faces
cluster01 = Cluster.ByTopologies([ap01, doors01])
# Vertices from cluster and copy them to avoid modifying the original topology when creating the graph
vertices01 = Topology.Vertices(cluster01)
new_vertices01 = [Topology.Copy(v) for v in vertices01]
# Edges from cluster and copy them to avoid modifying the original topology when creating the graph
edges01 = Topology.Edges(cluster01)
new_edges01 = [Topology.Copy(e) for e in edges01]
#Primal Graph from vertices and edges
primal01 = Graph.ByVerticesEdges(new_vertices01, new_edges01)

# =========== APARTMENT 02 ============
# Cluster from cell complex and faces
cluster02 = Cluster.ByTopologies([ap02, doors02])
# Vertices from cluster and copy them to avoid modifying the original topology when creating the graph
vertices02 = Topology.Vertices(cluster02)
new_vertices02 = [Topology.Copy(v) for v in vertices02]
# Edges from cluster and copy them to avoid modifying the original topology when creating the graph
edges02 = Topology.Edges(cluster02)
new_edges02 = [Topology.Copy(e) for e in edges02]
#Primal Graph from vertices and edges
primal02 = Graph.ByVerticesEdges(new_vertices02, new_edges02)
```

## 3.2. Assign Visual Attributes

```python
# =========== APARTMENT 01 ============
for v in new_vertices01:
    d = Dictionary.ByKeysValues(["size", "color"], [6, "blue"])
    v = Topology.SetDictionary(v, d)
for e in new_edges01:
    d = Dictionary.ByKeysValues(["width", "color"], [2, "blue"])
    e = Topology.SetDictionary(e, d)
    
# =========== APARTMENT 02 ============
for v in new_vertices02:
    d = Dictionary.ByKeysValues(["size", "color"], [6, "red"])
    v = Topology.SetDictionary(v, d)
for e in new_edges02:
    d = Dictionary.ByKeysValues(["width", "color"], [2, "red"])
    e = Topology.SetDictionary(e, d)
```

## 3.3. Show the graph

```python
Topology.Show(primal01, primal02,
              vertexSizeKey="size",
              vertexColorKey="color",
              edgeWidthKey="width",
              edgeColorKey="color",
              faceOpacity=0.3,
              backgroundColor="white",
              width=500,
              height=500,
              renderer=renderer)
```

## 4.1. Derive Dual Graph

```python
dual01 = Graph.ByTopology(ap01, direct=True)
dual02 = Graph.ByTopology(ap02, direct=True)
```

## 4.2. Assign visual attributes to the vertices and edges of the graph

```python
# =========== APARTMENT 01 ============
vertices01 = Graph.Vertices(dual01)
for v in vertices01:
    d = Dictionary.ByKeysValues(["size", "color"], [12, "blue"])
    v = Topology.SetDictionary(v, d)
edges01 = Graph.Edges(dual01)
for e in edges01:
    d = Dictionary.ByKeysValues(["width", "color"], [2, "blue"])
    e = Topology.SetDictionary(e, d)
    
# =========== APARTMENT 02 ============
vertices02 = Graph.Vertices(dual02)
for v in vertices02:
    d = Dictionary.ByKeysValues(["size", "color"], [12, "red"])
    v = Topology.SetDictionary(v, d)
edges02 = Graph.Edges(dual02)
for e in edges02:
    d = Dictionary.ByKeysValues(["width", "color"], [2, "red"])
    e = Topology.SetDictionary(e, d)
```

## 4.3. Show the geometry and graph

```python
Topology.Show(dual01, dual02,ap01, ap02,
              vertexSizeKey="size",
              vertexColorKey="color",
              edgeWidthKey="width",
              edgeColorKey="color",
              faceOpacity=0.3,
              backgroundColor="white",
              width=500,
              height=500,
              renderer=renderer)
```
